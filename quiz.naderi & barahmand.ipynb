```python
#5)
def isprime(n):
    '''
    check if the number is prime
    '''

    if n <= 1: 

        return False

    for i in range(2, n): 

        if n % i == 0: 

            return False

    return True

n = int(input('please enter your number: '))

while True:

    if isprime(n):

        new = n + 1

        while True:

            new += 1

            if isprime(new):
            
                print(new)

                break

        break

    n = int(input('enter another number: '))
```
