```python
def find_doc(x):
    
    if x.split('.')[1] == 'doc':
        
        return x

def find_docx(x):
    
    if x.split('.')[1] == 'docx':
        
        return x

def find_txt(x):
    
    if x.split('.')[1] == 'txt':
        
        return x

def find_exe(x):
    
    if x.split('.')[1] == 'exe':
        
        return x


lst = []

num = int(input('how many text files do you have? (doc, docx, txt, exe)'))

for i in range(0, num):

    ele = input()
 
    lst.append(ele)

list(map(lambda y : y.split('.')[1], lst))

print(list(filter(find_exe, lst)), list(filter(find_doc, lst)), list(filter(find_docx, lst)), list(filter(find_txt, lst)))
```
