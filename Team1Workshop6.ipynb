```python
def euclid_distance(pi , pf) :
    
        from math import pow , sqrt

        total_squares = 0

        for i , valuepi in enumerate(pi) :
     
            for f , valuepf in enumerate(pf) :
            
                if i == f :
            
                    total_squares += pow((valuepf - valuepi) , 2)
            
        distance = sqrt(total_squares)

        return distance

def manhattan_distance(pi , pf) :
    
    distance = 0

    for i , valuepi in enumerate(pi) :
     
        for f , valuepf in enumerate(pf) :
            
            if i == f :
            
                distance += abs(valuepf - valuepi)

    return distance

import numpy as np

import matplotlib.pyplot as plt

import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']

dataset = pd.read_csv(url, names=names)
    
px = []

print("Choosen observation :\n" , dataset.iloc[0] , "\n")

print(39 * " " , "euclide-distance" , 5 * " " , "manhattan-distance" , "\n")
    
for j in range(4) :
        
    px.append(dataset.iloc[0 , j])
        
px = tuple(px)

for i in range(1 , len(dataset)) :
    
    datapoint = []
    
    for j in range(4) :
        
        print(dataset.iloc[i][j] , end = "       ")
         
        datapoint.append(dataset.iloc[i][j])
        
    datapoint = tuple(datapoint) 
    
    print(euclid_distance(px , datapoint) , end = "       ")
    
    print(round(manhattan_distance(px , datapoint),1))
    
    print("\n")
    
    
    
```
